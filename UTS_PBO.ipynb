```python
# ============================================================
# INHERITANCE - Sistem Rekam Medis Digital
# ============================================================

# Kelas induk (parent class) - berisi atribut dasar pengguna
class Pengguna:
    def __init__(self, nama, umur, alamat):
        self.nama = nama        # nama lengkap pengguna
        self.umur = umur        # umur pengguna
        self.alamat = alamat    # alamat tempat tinggal

    # Method umum yang bisa dipakai semua turunan
    def info(self):
        print(f"Nama    : {self.nama}")
        print(f"Umur    : {self.umur} tahun")
        print(f"Alamat  : {self.alamat}")


# Kelas turunan Dokter - mewarisi semua atribut Pengguna
class Dokter(Pengguna):
    def __init__(self, nama, umur, alamat, spesialisasi):
        super().__init__(nama, umur, alamat)  # panggil constructor induk
        self.spesialisasi = spesialisasi      # atribut khusus dokter

    def info(self):
        super().info()   # tampilkan info dasar dari induk
        print(f"Spesialisasi : {self.spesialisasi}")
        print(f"Peran        : Dokter")


# Kelas turunan Pasien - mewarisi semua atribut Pengguna
class Pasien(Pengguna):
    def __init__(self, nama, umur, alamat, keluhan):
        super().__init__(nama, umur, alamat)  # panggil constructor induk
        self.keluhan = keluhan                # atribut khusus pasien

    def info(self):
        super().info()   # tampilkan info dasar dari induk
        print(f"Keluhan  : {self.keluhan}")
        print(f"Peran    : Pasien")


# ============================================================
# PENGUJIAN
# ============================================================
print("===== DATA DOKTER =====")
dokter1 = Dokter("dr. Sari", 35, "Jogja", "Umum")
dokter1.info()

print()
print("===== DATA PASIEN =====")
pasien1 = Pasien("Rizal", 22, "Klaten", "Demam & Batuk")
pasien1.info()
```

```python
# ============================================================
# POLIMORFISME - Sistem Rekam Medis Digital
# ============================================================

# Kelas induk dengan method hitung() sebagai template
class BiayaMedis:
    def __init__(self, biaya_dasar):
        self.biaya_dasar = biaya_dasar  # biaya awal sebelum perhitungan

    def hitung(self):
        # method ini akan di-override oleh kelas turunan
        pass

    def tampilkan(self):
        hasil = self.hitung()  # panggil hitung() - polymorfik!
        print(f"Biaya Dasar  : Rp {self.biaya_dasar:,.0f}")
        print(f"Total Bayar  : Rp {hasil:,.0f}")


# Turunan 1 - menghitung Pajak (10%)
class HitungPajak(BiayaMedis):
    def hitung(self):
        pajak = self.biaya_dasar * 0.10   # pajak 10% dari biaya dasar
        return self.biaya_dasar + pajak   # total = biaya + pajak


# Turunan 2 - menghitung Diskon (20%)
class HitungDiskon(BiayaMedis):
    def hitung(self):
        diskon = self.biaya_dasar * 0.20  # diskon 20% dari biaya dasar
        return self.biaya_dasar - diskon  # total = biaya - diskon


# Turunan 3 - menghitung Asuransi (ditanggung 50%)
class HitungAsuransi(BiayaMedis):
    def hitung(self):
        tanggungan = self.biaya_dasar * 0.50  # asuransi menanggung 50%
        return self.biaya_dasar - tanggungan  # pasien hanya bayar sisanya


# ============================================================
# PENGUJIAN - Tombol "Hitung" yang sama, hasil berbeda
# ============================================================
tagihan_list = [
    HitungPajak(500_000),       # pasien umum, kena pajak
    HitungDiskon(500_000),      # pasien lansia, dapat diskon
    HitungAsuransi(500_000),    # pasien BPJS, ditanggung asuransi
]

label = ["Pajak", "Diskon", "Asuransi"]

for i, tagihan in enumerate(tagihan_list):
    print(f"===== Perhitungan {label[i]} =====")
    tagihan.tampilkan()
    print()
```

```python
# ============================================================
# ENKAPSULASI - Sistem Rekam Medis Digital
# ============================================================

class RekamMedis:
    def __init__(self, nama_pasien, pin, diagnosa, harga_pokok):
        self.nama_pasien = nama_pasien      # atribut publik - bisa diakses bebas

        # Atribut PRIVATE - diawali __ agar tersembunyi dari luar
        self.__pin = pin                    # PIN rahasia pasien
        self.__diagnosa = diagnosa          # diagnosa bersifat rahasia medis
        self.__harga_pokok = harga_pokok    # harga pokok layanan, tidak dipublikkan

    # GETTER - cara aman untuk membaca data rahasia
    def get_diagnosa(self, pin_input):
        if pin_input == self.__pin:         # verifikasi PIN dulu
            return self.__diagnosa          # baru tampilkan diagnosa
        else:
            return "Akses ditolak! PIN salah."

    def get_harga_pokok(self, pin_input):
        if pin_input == self.__pin:         # verifikasi PIN
            return f"Rp {self.__harga_pokok:,.0f}"
        else:
            return "Akses ditolak! PIN salah."

    # SETTER - cara aman untuk mengubah data rahasia
    def set_diagnosa(self, pin_input, diagnosa_baru):
        if pin_input == self.__pin:         # hanya dokter dengan PIN yang bisa ubah
            self.__diagnosa = diagnosa_baru
            print("Diagnosa berhasil diperbarui.")
        else:
            print("Gagal! PIN tidak valid.")

    # Menampilkan info publik saja (tanpa data rahasia)
    def info_publik(self):
        print(f"Nama Pasien  : {self.nama_pasien}")
        print(f"PIN          : [TERSEMBUNYI]")
        print(f"Diagnosa     : [RAHASIA - butuh PIN]")
        print(f"Harga Pokok  : [RAHASIA - butuh PIN]")


# ============================================================
# PENGUJIAN
# ============================================================
rekam = RekamMedis("Laisya", "1234", "Hipertensi", 350_000)

print("===== INFO PUBLIK =====")
rekam.info_publik()

print()
print("===== AKSES DENGAN PIN SALAH =====")
print("Diagnosa  :", rekam.get_diagnosa("0000"))     # ditolak

print()
print("===== AKSES DENGAN PIN BENAR =====")
print("Diagnosa  :", rekam.get_diagnosa("1234"))     # diterima
print("Harga     :", rekam.get_harga_pokok("1234"))  # diterima

print()
print("===== UPDATE DIAGNOSA =====")
rekam.set_diagnosa("1234", "Hipertensi + Diabetes")  # berhasil diubah
print("Diagnosa baru:", rekam.get_diagnosa("1234"))
```
